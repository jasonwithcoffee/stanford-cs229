```python
def get_words(message):
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    
    message = message.lower()
    for ele in message:  
        if ele in punc:  
            message = message.replace(ele, "")
    message = message.split(sep=" ") 
    
    return list(filter(None,message))
```
